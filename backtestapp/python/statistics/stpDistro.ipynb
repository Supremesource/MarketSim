```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
currentP = 100
liqP = 200
priceArea = abs(currentP - liqP)
chunk = priceArea / 100  

# Define the intervals and their corresponding values
intervals = np.arange(1, 101, 10)
values = currentP + chunk * np.linspace(5, 50, 10)

# Plot the distribution
plt.bar(intervals, values, width=9, align='edge')
plt.xlabel('Randomizer value')
plt.ylabel('Final price')
plt.title('Distribution of final prices')
plt.xticks(intervals)
plt.grid(True)
plt.show()
```
