```python
import matplotlib.pyplot as plt
import numpy as np

# Define the leverage thresholds, flipping the key-value pairs and their order
thresholds = {
    500: 1,
    470: 2,
    465: 3,
    460: 4,
    455: 5,
    450: 6,
    445: 7,
    440: 8,
    435: 9,
    430: 10,
    425: 20,
    420: 30,
    415: 40,
    410: 50,
    405: 60,
    400: 70,
    395: 80,
    390: 90,
    305: 100
}

# Calculate probabilities as the proportion of the total range (500) that is below each threshold
probabilities = [value / 500 for value in thresholds.keys()]

# Plot the curve
plt.plot(probabilities, list(thresholds.values()))
plt.title('Leverage Distribution Curve')
plt.xlabel('Probability')
plt.ylabel('Leverage')
plt.grid(True)

# Customize x-axis labels
labels = ['Smallets', 'Biggest']
plt.xticks([min(probabilities), max(probabilities)], labels)

# Invert y-axis
plt.gca().invert_yaxis()


plt.show()
```
